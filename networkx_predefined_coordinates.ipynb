```python
import random
import networkx as nx
import matplotlib.pyplot as plt
```

## Example-1: 5 nodes

```python
nodes = ["X1", "X2", "X3", "X4", "X5"]
p = len(nodes)

# Assign coordinates to each node
node_pos = {'X1':[0, 0],
            'X2':[0, 1],
            'X3':[0.5, 0.5],
            'X4':[1, 0],
            'X5':[1, 1]}
node_pos
```

```python
plt.figure(figsize = (2, 2))
g_rd = nx.gn_graph(p, seed = 1)
g = nx.DiGraph()
g.add_nodes_from(nodes)

for node_from, node_to in g_rd.edges():
    g.add_edge(nodes[node_from], nodes[node_to])

# Visualize the network using predefined node coordinates
nx.draw(g, pos = node_pos, with_labels = True, node_color='lightblue')
```

## Example-2: 50 nodes

```python
import numpy as np
```

```python
nodes = [f"X{i}" for i in range(1, 51)]
p = len(nodes)

# Assume there are five clusters, manually specify the cluster center coordinates
center_coordinates = [[0.2, 0.2], [0.8, 0.2], [0.5, 0.5], [0.2, 0.8], [0.8, 0.8]]
num_nodes_per_cluster = 10


# Assign coordinates to each node
nodes_coordinates = []
for center in center_coordinates:
    for i in range(num_nodes_per_cluster):
        nodes_coordinates.append(center + 0.05 * np.random.randn(2))

node_pos = dict(zip(nodes, nodes_coordinates))

print(node_pos["X1"], node_pos["X6"], node_pos["X11"], node_pos["X16"], node_pos["X21"])
```

```python
plt.figure(figsize = (12, 4))
g_rd = nx.gn_graph(p, seed = 1)
g = nx.DiGraph()
g.add_nodes_from(nodes)

for node_from, node_to in g_rd.edges():
    g.add_edge(nodes[node_from], nodes[node_to])

# Visualize the network using predefined node coordinates
nx.draw(g, pos = node_pos, with_labels = True, node_color='lightblue')
```
